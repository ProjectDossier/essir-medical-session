#  Load Queries

```python
# Queries
dataset = ir_datasets.load("clinicaltrials/2021/trec-ct-2022")
queries = pd.DataFrame(dataset.queries_iter())
del dataset
display(queries)
```

# KeyBERT

https://github.com/MaartenGr/KeyBERT#embeddings

```python
kw_model = KeyBERT()

def keyBert_extraction(query,kw_model):
  keywords_list = kw_model.extract_keywords(query, keyphrase_ngram_range=(1, 2), stop_words='english',
                              use_maxsum=False, nr_candidates=25, top_n=20)
  # display(keywords_list)
  # Extract keywords from the list
  keywords = [item[0] for item in keywords_list]

  # Join keywords into a single text input
  text_input = ', '.join(keywords)

  return (text_input)


queries['keybert_topics'] = queries['text'].apply(lambda x: keyBert_extraction(x,kw_model))

display(queries)
```

```python
from flair.embeddings import TransformerDocumentEmbeddings

roberta = TransformerDocumentEmbeddings('roberta-base')
kw_model = KeyBERT(model=roberta)

queries['keybert_topics2'] = queries['text'][:2].apply(lambda x: keyBert_extraction(x,kw_model))

display(queries)
```
